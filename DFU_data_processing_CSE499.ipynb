<h2>This Converts CSV xmin,ymin,xmax,ymax to YOLO normalized & Saves to txt</h2>

```python
for i in range(0, len(df)):
    label = df.iloc[i]['name']
    label = label[:6] # This drops '.jpg' from filename
    xmin = df.iloc[i]['xmin']
    ymin = df.iloc[i]['ymin']
    xmax = df.iloc[i]['xmax']
    ymax = df.iloc[i]['ymax']
   
    #print("label: " +label+ " xmin: " +xmin+ " ymin: " +ymin+ " xmax: " +xmax+ " ymax: " +ymax)
    
    # (xmin, ymin, xmax, ymax) to (class, x, y, width,height)
    xcen = float((xmin + xmax)) / 2 
    ycen = float((ymin + ymax)) / 2 

    w = float((xmax - xmin)) 
    h = float((ymax - ymin)) 

    # Here we are normalizing for yolo.
    # All images in DFU are 640x480
    xcen = xcen / 640
    ycen = ycen / 480
    w = w / 640
    h = h / 480
    
    xcen = str(xcen)
    ycen = str(ycen)
    w = str(w)
    h = str(h)
    
    #print("0 " +xcen+ " " +ycen+ " " +w+ " " +h)
    
    # If multiple annotaion exist in a single file then write annontaion in newline into the current file
    # Else create a new file and write annotation

    if os.path.isfile(label+'.txt'):
        print('multiple file: '+label)
        with open(label+'.txt', 'a') as fd:
            fd.write(f'\n{"0 " +xcen+ " " +ycen+ " " +w+ " " +h}')
    else:
        with open(label+'.txt', 'x') as f:
            f.write("0 " +xcen+ " " +ycen+ " " +w+ " " +h)
    
```

<h2>This Part Renames All Files From 0 to N</h2>

```python
import os
from os import path
import shutil

Source_Path = 'D:\CSE 499\DFU Dataset\Labels'
Destination = Source_Path
#dst_folder = os.mkdir(Destination)


def main():
    for count, filename in enumerate(os.listdir(Source_Path)):
        dst =  str(count) + ".txt"

        # rename all the files
        os.rename(os.path.join(Source_Path, filename),  os.path.join(Destination, dst))


# Driver Code
if __name__ == '__main__':
    main()
```

<h2>This Converts xmin, ymin, xmax, ymax txt files to YOLO format txt files</h5>

```python
import os
textLocation = 'Labels/'
textDestination = 'Augmented Label/YOLOFormat/Original/'

from tqdm import tqdm
for i in tqdm(range(2000)):
    newTextLocation = textLocation + str(i) + '.txt'
    textLabel = textDestination + str(i) + '.txt'

    f = open(newTextLocation, "r")
    line_count = 0 
    for line in f:
        if line != "\n":
            line_count += 1

    #print("Total Line:",line_count) 
    f.close() # Annotation Count Ends

    f = open(newTextLocation, "r")
    while line_count > 0:
        values = f.readline()
        splitted = values.split()
        xmin = float(splitted[0])
        ymin = float(splitted[1])
        xmax = float(splitted[2])
        ymax = float(splitted[3])

        # (xmin, ymin, xmax, ymax) to (class, x, y, width,height)
        xcen = float((xmin + xmax)) / 2 
        ycen = float((ymin + ymax)) / 2 

        w = float((xmax - xmin)) 
        h = float((ymax - ymin)) 

        # Here we are normalizing for yolo.
        # All images in DFU are 640x480
        xcen = xcen / 640
        ycen = ycen / 480
        w = w / 640
        h = h / 480
    
        xcen = str(xcen)
        ycen = str(ycen)
        w = str(w)
        h = str(h)
    
        #print("0 " +xcen+ " " +ycen+ " " +w+ " " +h)
        line_count -= 1
    
        # If multiple annotaion exist in a single file then write annontaion in newline into the current file
        # Else create a new file and write annotation

        if os.path.isfile(textLabel):
            #print('multiple file: '+label)
            with open(textLabel, 'a') as fd:
                fd.write(f'\n{"0 " +xcen+ " " +ycen+ " " +w+ " " +h}')
        else:
            with open(textLabel, 'x') as fd1:
                fd1.write("0 " +xcen+ " " +ycen+ " " +w+ " " +h)
    f.close()


```
